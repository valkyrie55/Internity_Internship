Heatmap: 

A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors.It is often desirable to show data which depends on two independent variables as a color coded image plot. If the data is categorical, this would be called a categorical heatmap

We need to choose a relevant colour palette, use cluster analysis and thus permute the rows and the columns of the matrix to place similar values near each other according to the clustering.

* It is really useful to display a general view of numerical data,not to extract specific data point.
* It is a very good way to perform exploratory data analysis.
* Heatmaps can reveal general pattern in the dataset, thus widely used in data analysis in python

```python
#Creating a heatmap of sales of local shopkeepers
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

Products = ["oil", "soap", "snacks", "cold drinks",
              "clothes", "wheat", "lotion"]
shopkeepers = ["Jonathan ", "sheela", "Adam",
           "clit", "Shawn", "Nicola", "Eric"]

sales = np.array([[0.3, 1.4, 3.5, 3.5, 2.0, 2.3, 1.0],
                    [2.3, 1.0, 1.0, 0.5, 2.1, 3.5, 5.2],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(sales)

# We want to show all ticks...
ax.set_xticks(np.arange(len(shopkeepers)))
ax.set_yticks(np.arange(len(Products)))
ax.set_xticklabels(shopkeepers)
ax.set_yticklabels(Products)

#set the alignment of the tick labels 
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

#create heatmap
for i in range(len(Products)):
    for j in range(len(shopkeepers)):
        text = ax.text(j, i, sales[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Sales of local shopkeepers")
fig.tight_layout()
plt.show()
```
