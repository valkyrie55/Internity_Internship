```python
df = df.drop(['CHAS','DIS','PTRATIO'],axis=1)
#Drop unnecessary columns
```

```python
plt.bar(df.ZN,df.CRIM)
plt.xlabel('proportion of residential land zoned for lots over 25,000 sq.ft.')
plt.ylabel('per capita crime rate by town')
plt.title('proportion of residential land vs per capita crime rate by town')
plt.plot()
#increase in residental land cause decrease in crime rate
```

```python
plt.bar(df.ZN,df.AGE)
plt.xlabel('proportion of residential land')
plt.ylabel('Proportion of owner-occupied units')
plt.plot()
#with increase in proportion of residential land,Proportion of owner-occupied units is increasing
```
